```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars_data=pd.read_csv(r"C:\computer science\nptel\Toyota.csv",index_col=0,na_values=["??","????"])
```

```python
cars_data
```

```python
#remove missing values
cars_data.dropna(axis=0,inplace=True)
```

```python
cars_data.shape
```

```python
#scatter plot in sea born
```

```python
#price vs age
```

```python
sns.set(style="darkgrid")
sns.regplot(x=cars_data['Age'],y=cars_data['Price'])
```

```python
#scatter plot without regression fit line
sns.regplot(x=cars_data['Age'],y=cars_data['Price'],fit_reg=False)
```

```python
#change markers
sns.regplot(x=cars_data['Age'],y=cars_data['Price'],marker="*",fit_reg=False)
```

```python
#scatter plot of price vs age by fueltype----hue is used
sns.lmplot(x='Age',y='Price',data=cars_data,fit_reg=False,hue='FuelType',legend=True,palette="Set1")
```
